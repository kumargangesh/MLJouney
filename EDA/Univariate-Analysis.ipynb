Univariate Analysis


Here we are going to understand Univariate Analysis,
here Univariate Analysis means that we are going to analyze the single column

1. Categorical data: like pclass, survived, gender

```python
import seaborn as sb
```

1. Countplot from seaborn

```python
sb.countplot(df['Survived'])
```

2. PiePlot

```python
df['Survived'].value_counts().plot(kind = "pie", autopct="%.2f")
```

2. Numerical Data: like age, fare

1. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age']) # basically here we are creating a histogram of age
```

2. Distplot

```python
sb.distplot(df['Age'])
```

3. Boxplot

```python
sb.boxplot(df['Age'])
```
