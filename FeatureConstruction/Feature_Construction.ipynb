#**Feature Construction**

```python
new_data
```

```python
new_df = pd.DataFrame(new_data, columns=['Survived', 'Pclass', 'Age', 'SibSp', 'Parch'])
new_df.head()
```

```python
x = new_df.iloc[:, 1:]
y = new_df['Survived']
```

```python
np.mean(cross_val_score(LR(), x, y, scoring='accuracy', cv=20))
```

```python
new_df['Family_Type'] = new_df['SibSp'] + new_df['Parch'] + 1
new_df.head()
```

```python
def members(a):
  if a == 1:
    return 0
  elif a > 1 and a <= 4:
    return 1
  else:
    return 2
```

```python
new_df['Family_Size'] = new_df['Family_Type'].apply(members)
new_df.head()
```
