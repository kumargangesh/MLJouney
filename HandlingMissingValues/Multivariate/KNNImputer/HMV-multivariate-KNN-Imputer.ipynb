```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split as TTS

from sklearn.impute import SimpleImputer as SI, KNNImputer as KNNI
from sklearn.linear_model import LogisticRegression as LR

from sklearn.metrics import accuracy_score as acc
```

```python
data = pd.read_csv('train.csv', usecols=['Age', 'Fare', 'Pclass', 'Survived'])
data.head()
```

```python
data.isnull().mean()*100
```

```python
x = data.iloc[:, 1:]
y = data['Survived']
```

```python
trainx, testx, trainy, testy = TTS(x, y, test_size = 0.2, random_state=42)
```

#**Implementing KNN Imputer**

```python
knni = KNNI()

trainx_knni = knni.fit_transform(trainx)
testx_knni = knni.transform(testx)
```

#**training and tesing the model**

```python
lr = LR()

lr.fit(trainx_knni, trainy)

y_pred = lr.predict(testx_knni)

acc(testy, y_pred)*100
```

#**now try to handle missing values using SimpleImputer**

```python
si = SI()

trainx_si = si.fit_transform(trainx)
testx_si = si.transform(testx)
```

```python
lr = LR()

lr.fit(trainx_si, trainy)
predict = lr.predict(testx_si)

acc(testy, predict)*100
```
