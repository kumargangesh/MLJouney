```python
import pickle as pk
import pandas as pd
import numpy as np
```

```python
# opening or loading the pickle files

ohe_sex = pk.load(open('ohe_sex.pkl', 'rb'))
ohe_embarked = pk.load(open('ohe_embarked.pkl', 'rb'))
dtc = pk.load(open('model.pkl', 'rb'))
```

```python
# here we are providing the testing data

test_data = np.array([2, 'male', 31.0, 0, 0, 10.5, 'Q'], dtype=object).reshape(1, 7)
```

```python
test_data
```

```python
# here we are encoding the sex column from ohe_sex

test_data_sex = ohe_sex.transform(test_data[:, 1].reshape(1, 1))
```

```python
test_data_sex.toarray()
```

```python
# here we are encoding the embarked column from ohe_embak=rked

test_data_embarked = ohe_embarked.transform(test_data[:, -1].reshape(1, 1))
```

```python
test_data_embarked.toarray()
```

```python
# here we are fetching the age column form test_data

test_data_age = test_data[:, 2].reshape(1, 1)
```

```python
test_data_age
```

```python
test_data
```

```python
# here we are again concatenating the remainig columns of test_data, test_data_age, test_data, sex, test_data_embarked columns

test_data_transform = np.concatenate((test_data[:, [0, 3, 4, 5]], test_data_age, test_data_sex.toarray(), test_data_embarked.toarray()), axis = 1)
```
